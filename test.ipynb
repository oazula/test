
**Ejercicio de Simulación Monte Carlo para Estimar π**

El método de Monte Carlo es una técnica de simulación que se utiliza para estimar valores desconocidos mediante la generación de números aleatorios. En este ejercicio, vamos a utilizar el método de Monte Carlo para estimar el valor de π.

**Instrucciones:**

1. Genera un gran número de puntos aleatorios dentro de un cuadrado de lado 2 unidades, centrado en el origen (0, 0).
2. Contabiliza cuántos de estos puntos caen dentro de un círculo inscrito en el cuadrado, con radio 1 unidad.
3. Utiliza la proporción de puntos dentro del círculo con respecto al total de puntos generados para estimar el valor de π. La relación será aproximadamente igual a la relación entre el área del círculo y el área del cuadrado, que es π/4.

**Consejos para la ejecución:**

1. Asegúrate de entender cómo se generan los puntos aleatorios dentro del cuadrado. Utiliza la función `random.uniform()` de Python para generar coordenadas x e y aleatorias dentro del rango [-1, 1].
2. Para verificar si un punto está dentro del círculo, calcula la distancia euclidiana desde el origen hasta el punto `(x, y)` utilizando la fórmula `x**2 + y**2 <= 1`.
3. Grafica los puntos generados para visualizar cómo están distribuidos dentro y fuera del círculo. Utiliza matplotlib para crear un gráfico de dispersión y distingue entre los puntos dentro y fuera del círculo con diferentes colores.
4. Experimenta con diferentes cantidades de puntos generados (por ejemplo, 1000, 10000, 100000) para ver cómo afecta la precisión de la estimación de π.


```python
import random
import matplotlib.pyplot as plt

#funcion para generar datos aleatorios
def estimate_pi(num_points):

    points_inside_circle = 0
    points_inside_x = []
    points_inside_y = []
    points_outside_x = []
    points_outside_y = []
    
    for _ in range(num_points):
        # generacion como tal
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        
        # clsasificacion si esta dentro o fuera del circulo
        if x**2 + y**2 <= 1:
            points_inside_circle += 1
            points_inside_x.append(x)
            points_inside_y.append(y)
        else:
            points_outside_x.append(x)
            points_outside_y.append(y)
    
    # con la fraccion teorica calculamos pi: fracoion_teorica=Area_circulo/Area_cuadrado=pi*r^2/2*r*2*r=pi/4-->pi=4*fraccion_teorica
    pi_estimate = 4 * points_inside_circle / num_points
    
    # Plotting
    plt.figure(figsize=(8, 8))
    plt.scatter(points_inside_x, points_inside_y, color='blue', s=5, label='Points Inside Circle')
    plt.scatter(points_outside_x, points_outside_y, color='red', s=5, label='Points Outside Circle')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Monte Carlo Simulation for Estimating π\nNumber of Points: {num_points}, Estimated π: {pi_estimate:.6f}')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.legend()
    plt.show()
    
    return pi_estimate

#llamado
num_points = 10000  
estimated_pi = estimate_pi(num_points)
print(f"El valor estimado de pi usando {num_points} puntos: {estimated_pi}")
```


**Ejercicio: Cálculo del Factorial con Listas y Arrays**

**Descripción del Problema:**

Escribe un programa en Python para calcular el factorial de un número dado utilizando tanto listas como arrays. Compara el tiempo de ejecución de ambos métodos para demostrar la diferencia en velocidad.

**Instrucciones:**

1. Implementa una función `factorial_lista(n)` que calcule el factorial de un número entero positivo `n` utilizando una lista.
2. Implementa una función `factorial_array(n)` que calcule el factorial de un número entero positivo `n` utilizando un array de la librería `numpy`.
3. Prueba ambas funciones para un valor grande de `n` (por ejemplo, `n = 1000`) y mide el tiempo de ejecución de cada función.
4. Compara los tiempos de ejecución y observa cómo utilizar arrays de la librería `numpy` puede ser significativamente más rápido que usar listas para esta tarea.

**Consejos para la Solución:**

1. Utiliza un bucle `for` para iterar a través de los números desde 1 hasta `n` y calcular el factorial.
2. Cuando uses listas, inicializa una lista vacía para almacenar los valores del factorial y agrega cada valor a la lista dentro del bucle.
3. Cuando uses arrays, utiliza la librería `numpy` para crear un array y realiza multiplicación de elementos para calcular el factorial directamente.
4. Utiliza el módulo `time` en Python para medir el tiempo de ejecución de cada función.



```python
import numpy as np
import time

# Function to calculate factorial using lists
def factorial_list(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

# Function to calculate factorial using arrays
def factorial_array(n):
    result = np.arange(1, n+1)
    return np.prod(result)

# Test and compare execution times using %timeit
n = 4000

# Measure execution time for factorial_list


start_time = time.time()
factorial_list(n)
list_execution_time = time.time() - start_time
print("Execution time using lists: {:.10f} seconds".format(list_execution_time))

# Measure execution time for factorial_array

start_time = time.time()
factorial_array(n)
array_execution_time = time.time() - start_time
print("Execution time using arrays: {:.10f} seconds".format(array_execution_time))


```


**Ejercicio: Ajuste de Datos a una Función Exponencial**

**Enunciado del Problema:**

Se te proporciona un conjunto más grande de puntos de datos `(x, y)` que representan una relación desconocida entre dos variables. Tu tarea es ajustar estos puntos de datos a una función exponencial de la forma:

\[ y = a \cdot e^{bx} \]

donde `a` y `b` son los parámetros que se deben determinar.

**Instrucciones:**

1. Implementa una función en Python llamada `ajustar_exponencial(x_data, y_data)` que ajuste los puntos de datos dados a la función exponencial utilizando técnicas de ajuste de curvas.
2. Grafica los puntos de datos originales junto con la función exponencial ajustada para visualizar la relación entre las variables.
3. Imprime los valores de los parámetros `a` y `b` obtenidos del proceso de ajuste de curvas.
4. Opcionalmente, calcula la bondad del ajuste (por ejemplo, el valor R-cuadrado) para evaluar qué tan bien la función exponencial se ajusta a los datos.

**Consejos:**

- Utiliza la función `scipy.optimize.curve_fit()` para realizar el ajuste de curvas. Esta función requiere una suposición inicial para los parámetros `a` y `b`.
- Es posible que necesites importar bibliotecas necesarias como `numpy` y `scipy`.
- Visualiza los puntos de datos originales y la función exponencial ajustada utilizando matplotlib.
- Experimenta con diferentes suposiciones iniciales para los parámetros para obtener el mejor ajuste.
- Puedes calcular el valor R-cuadrado utilizando métodos apropiados de bibliotecas como `scipy` o implementando el cálculo tú mismo.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the exponential function
def exponential_func(x, a, b):
    return a * np.exp(b * x)

# Function to fit data to exponential function
def fit_exponential(x_data, y_data):
    # Perform curve fitting
    popt, pcov = curve_fit(exponential_func, x_data, y_data)
    return popt

# Genracion de datos para ajustar
x_data = np.linspace(0, 5, 100)  # generamos 100 puntos entre 0 y 5
y_data = 2 * np.exp(0.5 * x_data) + np.random.normal(0, 0.5, size=x_data.shape)  # calculamos el valor de y con la ecuacion exponencial

# ajuste
parameters = fit_exponential(x_data, y_data)
a, b = parameters

# grafica
plt.scatter(x_data, y_data, label='Original Data')
plt.plot(x_data, exponential_func(x_data, a, b), color='red', label='Fitted Exponential Function')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fitting Data to Exponential Function')
plt.legend()
plt.grid(True)
plt.show()

# los parametros
print("Fitted parameter a:", a)
print("Fitted parameter b:", b)
```
